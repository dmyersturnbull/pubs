```python
import pickle
import numpy as np
import seaborn as sns
import pandas
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pandas.read_csv('growth_data.csv', header=0) # , index_col=0

df['time'] = [float(t.replace(':', '.')) for t in df['time']]
df['OD600'] = [math.log(od) / math.log(2) for od in df['OD600']]
df['time'] = [t - 13.39 for t in df['time']]
#mu = u'\u03BC'
#df.columns = ['incubation time (hr)', 'log2 (OD600)', 'concentration (' + mu + 'L)']
```

```python
sns.lmplot('incubation time (hr)', 'log2 (OD600)', data=df, hue='concentration (' + mu + 'L)', fit_reg=True, size=10, scatter_kws={"s": 100})
sns.set(font_scale=2)
sns.plt.xlim(0,)
plt.show()
#plt.savefig("conc.png")
```
